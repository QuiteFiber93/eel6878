```python
import networkx as nx
import pandas as pd

chars = pd.read_csv("characters.csv", index_col="id")
edges = pd.read_csv("relations_weighted.csv")

G = nx.from_pandas_edgelist(edges, "source", "target",
                             edge_attr=["weight", "type"])
nx.set_node_attributes(G, chars["name"].to_dict(), "name")
```
